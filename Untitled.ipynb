**Тестовое задание:** необходимо составить программу на python, рисующую графики ошибок 1го и 2го рода для двух статистических гипотез (H0 и H1). Параметры распределения гипотез выбрать самостоятельно. Ссылку на код программы можно прикреплять к отклику

```python
# Преамбула
%matplotlib inline

import math
import matplotlib.pyplot as plt
import numpy as np
```

Имеется выборка объема n=900 из нормального распределения ${\mathsf N}_{a,1}$ и две простые гипотезы о среднем: $H_0=\{a=0\}$ и $H_1=\{a=2\}$. Рассмотрим следующий критерий (при некотором вещественном $c$):  

$\delta(X)$ = $H_{0}$, если X <= c  
              $H_{1}$, если X > c

Изобразим на графике соответствующие гипотезам плотности и вероятности ошибок первого и второго рода критерия $\delta$:

${α}$=${P}_{H_0}(\delta(X) = H_1)$ = ${P}_{H_0}(X>c)$

${β}$=${P}_{H_1}(\delta(X) = H_0)$ = ${P}_{H_1}(X<=c)$

```python
X = np.arange(-4,5,0.01)
```

```python
def f(X,m,sigma): #нормальное распределение
    density = []
    for i in range(0,len(X)):
        density.append( 1/(sigma*math.sqrt(2*math.pi) ) * math.exp(-(X[i]-m)**2/(2*sigma**2)) )  
    return density
```

```python
y1 = f(X,0,1)
y2 = f(X,2,1)
# Вспомогательная прямая для наглядности ошибок 1-го и 2-го рода
t = np.arange(0,max(f(X,0,1)),0.01)
c = [0.5 for i in range(len(t))]
```

```python
fig = plt.figure()
ax=plt.subplot(111)
ax.plot(X, y1, label='$N_{0,1}$')
ax.plot(X, y2, label='$N_{2,1}$')
ax.plot(c, t)
ax.fill_between(X,y2,t[0], where=X<c[0],color='c')
ax.fill_between(X,y1,t[0], where=X>=c[0],color='violet')
plt.xlabel('Исследуемая случайная величина')
plt.ylabel('Плотность вероятности')
plt.title('Плотности распределения для нулевой и альтернативной гипотез')
plt.text(-0.3, 0.33, r'$H_{0}$', fontsize = 13)
plt.text(1.8, 0.33, r'$H_{1}$', fontsize = 13)
plt.text(0.9, 0.04, r'$α$', fontsize = 13)
plt.text(0, 0.04, r'$β$', fontsize = 13)
plt.text(1.6, 0.2, r'$1-β$', fontsize = 13)
plt.text(-1, 0.2, r'$1-α$', fontsize = 13)
ax.legend()
plt.show()
```

Где:  
α - ошибка 1-го рода  
β - ошибка 2-го рода  
1-α - вероятность принятия правильного решения при истинной нулевой гипотезе  
1-β - вероятность принятия правильного решения при истинной альтернативной гипотезе
